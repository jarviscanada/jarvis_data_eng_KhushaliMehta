```python
import pandas as pd
```

```python
oo = pd.read_csv('olympics.csv', skiprows=4)
list(oo)
```

```python
jo =oo[oo.Athlete == 'OWENS, Jesse']
jo.Event.value_counts()
```

```python
gbm = oo[(oo.Medal == 'Gold') & (oo.Gender == 'Men') & (oo.Sport == 'Badminton')]
gbm.sort_values(by='Athlete')
```

```python
rec = oo[oo.Edition >= 1984]
rec.NOC.value_counts()
```

```python
gm100 = oo[(oo.Gender == 'Men') & (oo.Medal == 'Gold') & (oo.Event == '100m')]
gm100.sort_values('Edition', ascending=False)[['City', 'Edition', 'Athlete', 'NOC']]
```
